```python
import random
answer=random.randint(1, 100)
```

```python
answer
```

```python
user=int(input("추측값을 입력하시오"))
```

```python
count=1

while user != answer:
  if user<answer:
    count+=1
    print("더 큰 수를 입력하시오")
  else:
    count+=1
    print("더 작은 수를 입력하시오")
  user=int(input("추측값을 입력하시오"))
print(f"{count}번 만에 맞췄습니다")
```

```python
num=int(input("약수를 구할 수를 입력하시오: "))

for i in range(1,num+1):
  if num%i == 0:
    print(i)
```
