```python
# 입력한 분과 계산 결과(시간, 분)를 모두 출력

total_minutes = int(input("분을 입력하세요: "))

hours = total_minutes // 60
minutes = total_minutes % 60

print("입력한 분:", total_minutes)
print("계산 결과:", hours, "시간", minutes, "분")
```

```python
# 사용자에게 정수를 입력받아 짝수/홀수를 판별하는 프로그램
# 아이디어: 2로 나눈 나머지(%)가 0이면 짝수, 1이면 홀수입니다.

num = int(input("숫자를 입력하세요: "))

if num % 2 == 0:
    print(num, "는(은) 짝수입니다.")
else:
    print(num, "는(은) 홀수입니다.")
```

```python
num = int(input("숫자를 입력하세요: "))

if num > 0:
    print(num, "는(은) 양수입니다.")
else:
    print(num, "는(은) 음수입니다.")
```

```python
coffee_price=2000
latte_price=3000
greentea_price=2500
juice_price=2800

coffee_num=int(input('커피 주문수:'))
latte_num=int(input('라떼 주문수:'))
greentea_num=int(input('녹차 주문수:'))
juice_num=int(input('주스 주문수:'))

total_price=(coffee_price*coffee_num)+(latte_price*latte_num)+(greentea_price*greentea_num)+(juice_price*juice_num)

print(f'주문금액: {total_price}원')
```
