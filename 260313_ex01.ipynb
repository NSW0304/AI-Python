```python
# 두 수를 변수에 저장
a = 12
b = 5

# 사칙연산 결과를 각각 변수에 저장
add_result = a + b   # 덧셈
sub_result = a - b   # 뺄셈
mul_result = a * b   # 곱셈
div_result = a / b   # 나눗셈

# 출력문으로 결과 확인
print("a + b =", add_result)
print("a - b =", sub_result)
print("a * b =", mul_result)
print("a / b =", div_result)
```

```python
# 두 수를 입력받아 변수에 저장
a = int(input("첫 번째 정수를 입력하세요: "))
b = int(input("두 번째 정수를 입력하세요: "))

# 사칙연산 결과를 각각 변수에 저장
add_result = a + b    # 덧셈
sub_result = a - b    # 뺄셈
mul_result = a * b    # 곱셈
div_result = a / b    # 나눗셈(파이썬은 / 하면 실수 결과가 나올 수 있음)

# 출력문으로 결과 확인
print("a + b =", add_result)
print("a - b =", sub_result)
print("a * b =", mul_result)
print("a / b =", div_result)
```
