```python
orders = {}  # {"아메리카노": 2, "카페라떼": 1} 형태로 누적

print("[메뉴판]")
for name, price in cafe_menu.items():
    print(f"- {name}: {price:,}원")
print("\n메뉴명과 개수를 입력하세요. 종료하려면 '종료'를 입력하세요.\n")

while True:
    menu_name = input("메뉴명: ").strip()

    if menu_name == "종료":
        break

    if menu_name not in cafe_menu:
        print("없는 메뉴입니다. 메뉴판의 이름과 똑같이 입력해주세요.\n")
        continue

    qty_str = input("개수: ").strip()
    if not qty_str.isdigit() or int(qty_str) <= 0:
        print("개수는 1 이상의 정수로 입력해주세요.\n")
        continue

    qty = int(qty_str)
    orders[menu_name] = orders.get(menu_name, 0) + qty
    print(f"추가됨: {menu_name} {qty}개\n")

# 영수증 출력
print("\n[주문 내역]")
if not orders:
    print("주문이 없습니다.")
else:
    total_price = 0
    for name, qty in orders.items():
        unit_price = cafe_menu[name]
        item_price = unit_price * qty
        total_price += item_price
        print(f"- {name}: {qty}개 (개당 {unit_price:,}원) = {item_price:,}원")

    print(f"\n총 가격: {total_price:,}원")
```
