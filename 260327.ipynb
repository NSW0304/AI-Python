```python
score=int(input("점수:"))

if score >= 90:
  print("A")
elif 90>score>=80:
  print("B")
elif 80>score>=70:
  print("C")
else:
  print("D")
```

```python
import random

dice1=random.randint(1,6)
dice2=random.randint(1,6)

print(f"주사위1:{dice1}")
print(f"주사위2:{dice2}")

if dice1 == dice2:
  print("두 주사위의 눈은 같다.")
elif dice1 > dice2:
  print("주사위1의 눈이 주사위2의 눈보다 더 크다.")
else:
  print("주사위2의 눈이 주사위1의 눈보다 더 크다.")
```

```python
y=int(input("연도를 입력하시오:"))

if (y%400==0) or (y%4==0 and y%100!=0):
  print("윤년")
else:
  print("평년")
```

```python
#가위바위보

import random

man1 = random.randint(1,3)
man2 = random.randint(1,3)

if man1==1:
  print("사람1:가위")
elif man1==2:
  print("사람1:바위")
else:
  print("사람1:보")

if man2==1:
  print("사람2:가위")
elif man2==2:
  print("사람2:바위")
else:
  print("사람2:보")

if man1 == man2:
  print("결과:비김")
elif (man1 == 1 and man2 == 3) or (man1 == man2 +1):
  print("결과:사람1 승리")
else:
  print("결과:사람2 승리.")
```
